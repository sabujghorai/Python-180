```python
# ## Create an array containing the numbers from 1 to 10.
import numpy as np
arr = np.arange(0,11)
print("Array elements :",arr)
```

```python
# Find Shape, Dimensions and Total Elements
import numpy as np
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape of the array :",arr.shape) # (2,3) means 2 row 3 columns
print("Dimension of the array :",arr.ndim) # dimensions of the array is 2
print("Total elements are ;",arr.size) # total elements are 6
```

```python
# Reshape the array
import numpy as np

arr = np.arange(12)
print("Reshaped array :\n",arr.reshape(3,4))
```
