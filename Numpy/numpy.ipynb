## Numpy Basics

```python
import numpy as np
```

### Creating array from list

```python
array_1D = np.array([1,2,3,4,5])
print("1D array :", array_1D)

array_2d = np.array([[1,2,3],[4,5,6]])
print("2D array :", array_2d)
```

### List vs Numpy array

```python
np_list = [1,2,3]
print("Python list multiplication :",np_list*2)

np_array = np.array([4,5,6])
print("Python array Multiplication :" , np_array * 2) # Element wide Multiplication
```

### Creating Array from Scratch

```python
import numpy as np

zeros = np.zeros((3,4)) # 3 is the row and 4 is the column
print("Zeros array : \n",zeros)

ones = np.ones((2,3)) # 2 is the row and 3 is the column 
print("Ones array is :\n",ones)

full = np.full((2,2),7) # array with 2 row and 2 column all emenents are 7
print("Full array is : \n",full)

random = np.random.random((2,3)) # 2 rows and 3 column
print("Random array is :\n",random)

sequence = np.arange(0,10,2) # start, stop-1 ,step
print("Sequence array is : \n",sequence)
```

### Vector , Matrix and Tensor

```python
vector = np.array([1,2,3])
print("Vector is : ",vector)

matrix = np.array([[4,5,6],
                  [7,8,9]])
print("Matrix is :\n ",matrix)

tensor = np.array([[[1,2],[3,4]],
                   [[5,6],[7,8]]])
print("Tensor :\n",tensor)
```

### Array Properties

```python
array = np.array([[1,2,3],
                  [4,5,6]])
print("Shape :",array.shape)
print("Dimention :",array.ndim)
print("Size :",array.size)
print("Datatype :",array.dtype)
```

### Array Reshaping

```python
import numpy as np

arr = np.arange(12)
print("Original Array :",arr)

reshaped = arr.reshape((3,4)) # 3 row and 4 column
print("\n Reshaped array :\n",reshaped)

flattened = reshaped.flatten() # Arrange all the elements in a single line 
print("\n Flattened array :",flattened)

# ravel (returns view insted of copy)
ravel = flattened.ravel()
print("\n Ravel array :", ravel)

# Transpose
transpose = reshaped.T     # The dot T is a fnction to make the matrix transpose
print("Transpose matrix :\n",transpose)
```
